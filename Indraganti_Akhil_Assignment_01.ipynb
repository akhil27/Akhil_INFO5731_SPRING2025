```python
#This error occurs because the _initialize_csv method is referenced in the __init__ method but wasn't defined in the final code snippet.
#Let me fix this by adding the missing initialization methods. Here's the corrected implementation:


"""
This complete implementation includes:

Full Error Handling

Custom exceptions with meaningful messages

Validation for all input parameters

Graceful recovery from file errors

Atomic file operations to prevent corruption

Edge Case Coverage

Corrupted file detection and recovery

Write failure handling (disk full, permissions)

Duplicate product prevention

Negative value validation

Empty search term detection

Restocking limits enforcement

Key Features

CSV-based storage with automatic backups

UUID-based product identification

Inventory reports with financial breakdown

Automatic restocking system

Sample data generation

Maintenance Tools

CSV validation and repair

Automatic sample data creation

Backup system for corrupted files

Capacity management for stock levels

To use the system:

Create an InventoryManager instance

Use the provided methods to manage inventory

Handle specific edge cases with try/except blocks

Generate reports using generate_report()

The code includes a self-cleaning test example that demonstrates:

Error handling for various edge cases

Restocking functionality

Report generation

Clean test file management

All 10 specified edge cases are properly handled through validation checks, error raising, and recovery mechanisms.
"""

import csv
import os
import uuid
from datetime import datetime
from collections import defaultdict

class InventoryError(Exception):
    """Base class for inventory exceptions"""
    pass

class InventoryManager:
    MAX_CAPACITY = 1000  # Maximum stock capacity

    def __init__(self, filename='inventory.csv'):
        self.filename = filename
        self.fieldnames = ['ProductID', 'Name', 'Category', 'Price', 'Quantity', 'ReorderLevel']
        self._initialize_csv()
        self._validate_csv()
        self._generate_sample_data_if_needed()

    def _initialize_csv(self):
        """Create CSV file with headers if it doesn't exist"""
        if not os.path.exists(self.filename):
            try:
                with open(self.filename, 'w', newline='') as f:
                    writer = csv.DictWriter(f, fieldnames=self.fieldnames)
                    writer.writeheader()
            except IOError as e:
                raise InventoryError(f"Failed to initialize CSV: {str(e)}") from e

    def _validate_csv(self):
        """Validate CSV structure and data integrity"""
        try:
            with open(self.filename, 'r') as f:
                reader = csv.reader(f)
                headers = next(reader)
                if headers != self.fieldnames:
                    raise ValueError("Invalid CSV headers")

                for row_num, row in enumerate(reader, start=2):
                    if len(row) != len(self.fieldnames):
                        raise ValueError(f"Invalid row {row_num}: column count mismatch")
                    try:
                        float(row[3])  # Price
                        int(row[4])    # Quantity
                        int(row[5])    # ReorderLevel
                    except ValueError:
                        raise ValueError(f"Invalid data types in row {row_num}")
        except Exception as e:
            print(f"CSV validation failed: {e}")
            self._backup_csv()
            self._initialize_csv()

    def _backup_csv(self):
        """Create timestamped backup of corrupted file"""
        try:
            timestamp = datetime.now().strftime("%Y%m%d%H%M%S")
            backup_name = f"{self.filename}_backup_{timestamp}.csv"
            os.rename(self.filename, backup_name)
            print(f"Created backup: {backup_name}")
        except OSError as e:
            raise InventoryError(f"Backup failed: {str(e)}") from e

    def _generate_sample_data_if_needed(self):
        """Generate sample data if inventory is empty"""
        try:
            products = self._read_products()
            if len(products) < 50:
                categories = ['Electronics', 'Clothing', 'Books', 'Home', 'Toys']
                for i in range(50 - len(products)):
                    self.add_product(
                        name=f"Sample Product {i+1}",
                        category=categories[i % len(categories)],
                        price=round(10 + (i * 0.5), 2),
                        quantity=100,
                        reorder_level=20
                    )
        except InventoryError as e:
            print(f"Sample data generation failed: {str(e)}")

    def _read_products(self):
        """Read and validate products from CSV"""
        products = []
        if not os.path.exists(self.filename):
            return products

        with open(self.filename, 'r') as f:
            reader = csv.DictReader(f)
            for row in reader:
                try:
                    products.append({
                        'ProductID': row['ProductID'],
                        'Name': row['Name'],
                        'Category': row['Category'],
                        'Price': float(row['Price']),
                        'Quantity': int(row['Quantity']),
                        'ReorderLevel': int(row['ReorderLevel'])
                    })
                except (ValueError, KeyError):
                    print(f"Skipping invalid row: {row}")
        return products

    def _write_products(self, products):
        """Atomic write to CSV with error handling"""
        tempfile = self.filename + '.tmp'
        try:
            with open(tempfile, 'w', newline='') as f:
                writer = csv.DictWriter(f, fieldnames=self.fieldnames)
                writer.writeheader()
                writer.writerows(products)

            os.replace(tempfile, self.filename)
        except (PermissionError, OSError) as e:
            raise InventoryError(f"Failed to save inventory: {str(e)}") from e
        finally:
            if os.path.exists(tempfile):
                os.remove(tempfile)

    def add_product(self, name, category, price, quantity, reorder_level):
        """Add new product with validation"""
        products = self._read_products()

        # Check for duplicate names
        if any(p['Name'].lower() == name.lower() for p in products):
            raise InventoryError(f"Product '{name}' already exists")

        # Validate numerical values
        if price <= 0:
            raise InventoryError("Price must be positive")
        if quantity < 0:
            raise InventoryError("Quantity cannot be negative")
        if reorder_level < 0:
            raise InventoryError("Reorder level cannot be negative")

        product_id = str(uuid.uuid4())
        products.append({
            'ProductID': product_id,
            'Name': name,
            'Category': category,
            'Price': price,
            'Quantity': quantity,
            'ReorderLevel': reorder_level
        })

        self._write_products(products)
        return product_id

    def delete_product(self, product_id):
        """Delete product by ID"""
        products = self._read_products()
        original_count = len(products)
        products = [p for p in products if p['ProductID'] != product_id]

        if len(products) == original_count:
            raise InventoryError(f"Product {product_id} not found")

        self._write_products(products)

    def update_product(self, product_id, **kwargs):
        """Update product properties with validation"""
        products = self._read_products()
        found = False

        for product in products:
            if product['ProductID'] == product_id:
                found = True
                for key, value in kwargs.items():
                    if key not in self.fieldnames or key == 'ProductID':
                        raise InventoryError(f"Invalid field: {key}")

                    # Special validation for numeric fields
                    if key == 'Price':
                        if value <= 0:
                            raise InventoryError("Price must be positive")
                    elif key in ['Quantity', 'ReorderLevel']:
                        if value < 0:
                            raise InventoryError(f"{key} cannot be negative")

                    product[key] = value
                break

        if not found:
            raise InventoryError(f"Product {product_id} not found")

        self._write_products(products)
        return True

    def search_products(self, **kwargs):
        """Search products with validation"""
        # Validate search terms
        for key, value in kwargs.items():
            if key not in self.fieldnames:
                raise InventoryError(f"Invalid search field: {key}")
            if isinstance(value, str) and not value.strip():
                raise InventoryError("Empty search term provided")

        products = self._read_products()
        results = []

        for product in products:
            match = True
            for key, value in kwargs.items():
                if key not in product:
                    match = False
                    break
                if str(product[key]).lower() != str(value).lower():
                    match = False
                    break
            if match:
                results.append(product)

        return results

    def restock_products(self, restock_amount=50):
        """Restock products below reorder level"""
        try:
            products = self._read_products()
        except InventoryError as e:
            raise InventoryError("Cannot restock: Inventory file issues") from e

        updated = False
        for product in products:
            if product['Quantity'] < product['ReorderLevel']:
                new_quantity = product['Quantity'] + restock_amount

                # Enforce maximum capacity
                if new_quantity > self.MAX_CAPACITY:
                    new_quantity = self.MAX_CAPACITY

                if new_quantity != product['Quantity']:
                    product['Quantity'] = new_quantity
                    updated = True

        if updated:
            self._write_products(products)
        return updated

    def generate_report(self):
        """Generate inventory report"""
        products = self._read_products()
        category_breakdown = defaultdict(lambda: {'count': 0, 'value': 0.0})
        total_inventory_value = 0.0

        for product in products:
            product_value = product['Price'] * product['Quantity']
            total_inventory_value += product_value
            category_breakdown[product['Category']]['count'] += 1
            category_breakdown[product['Category']]['value'] += product_value

        report = {
            'total_value': total_inventory_value,
            'category_breakdown': dict(category_breakdown),
            'product_count': len(products)
        }

        if not products:
            report['message'] = 'No products in inventory'

        return report


# Example usage with edge case testing
if __name__ == "__main__":
    try:
        # Initialize with test file
        manager = InventoryManager('test_inventory.csv')

        # Test adding duplicate product
        try:
            manager.add_product("Sample Product 1", "Electronics", 10.99, 50, 20)
        except InventoryError as e:
            print(f"Duplicate add prevented: {e}")

        # Test invalid update
        test_id = list(manager.search_products(Name="Sample Product 1"))[0]['ProductID']
        try:
            manager.update_product(test_id, Quantity=-5)
        except InventoryError as e:
            print(f"Invalid update prevented: {e}")

        # Test empty search
        try:
            manager.search_products(Name="")
        except InventoryError as e:
            print(f"Empty search prevented: {e}")

        # Test restocking
        print("\nBefore restock:")
        print([p['Quantity'] for p in manager.search_products(Category="Electronics")][:3])
        manager.restock_products()
        print("After restock:")
        print([p['Quantity'] for p in manager.search_products(Category="Electronics")][:3])

        # Generate report
        report = manager.generate_report()
        print("\nInventory Report:")
        print(f"Total Value: ${report['total_value']:.2f}")
        print(f"Product Count: {report['product_count']}")
        for category, data in report['category_breakdown'].items():
            print(f"{category}: {data['count']} items, Value: ${data['value']:.2f}")

    except Exception as e:
        print(f"System error: {e}")
    finally:
        # Cleanup test file
        if os.path.exists('test_inventory.csv'):
            os.remove('test_inventory.csv')
```

```python
"""
Added input validation for flight/passenger details, case normalization to prevent duplicates (e.g., FL200 ≠ fl200), overbooking prevention, and enhanced error logging.
Introduced passenger tracking, atomic seat management, and utilization metrics in performance reports.
"""

import json
import logging

# Setup logging
logging.basicConfig(
    level=logging.INFO,
    format="%(asctime)s - %(levelname)s - %(message)s",
    force=True
)

class Passenger:
    def __init__(self, name, contact):
        if not name.strip() or not contact.strip():
            raise ValueError("Name/contact cannot be empty")
        self.name = name.strip().title()
        self.contact = contact.strip()

class Flight:
    def __init__(self, flight_number, origin, destination, total_seats):
        if total_seats <= 0:
            raise ValueError("Total seats must be > 0")

        self.flight_number = flight_number.strip().upper()
        self.origin = origin.strip().title()
        self.destination = destination.strip().title()
        self.total_seats = total_seats
        self.available_seats = total_seats
        self.reserved_seats = 0
        self.passengers = {}

    def to_dict(self):
        return {
            "flight_number": self.flight_number,
            "origin": self.origin,
            "destination": self.destination,
            "total_seats": self.total_seats,
            "available_seats": self.available_seats,
            "reserved_seats": self.reserved_seats,
            "passengers": {pid: p.__dict__ for pid, p in self.passengers.items()},
        }

    @classmethod
    def from_dict(cls, data):
        try:
            if data["total_seats"] <= 0:
                raise ValueError("Invalid seat count in saved data")
            return cls(
                data["flight_number"],
                data["origin"],
                data["destination"],
                data["total_seats"]
            )
        except KeyError as e:
            raise ValueError(f"Missing field: {str(e)}")

class FlightBookingSystem:
    def __init__(self):
        self.flights = {}
        self.load_flights()

    def load_flights(self):
        try:
            with open("flights.json", "r") as f:
                data = json.load(f)
                self.flights = {}
                for fid, fdata in data.items():
                    try:
                        flight = Flight.from_dict(fdata)
                        flight.available_seats = fdata["available_seats"]
                        flight.reserved_seats = fdata["reserved_seats"]
                        flight.passengers = {
                            pid: Passenger(p["name"], p["contact"])
                            for pid, p in fdata["passengers"].items()
                        }
                        self.flights[fid] = flight
                    except ValueError as e:
                        logging.error(f"Skipping invalid flight {fid}: {str(e)}")
        except (FileNotFoundError, json.JSONDecodeError):
            self.flights = {}

    def save_flights(self):
        with open("flights.json", "w") as f:
            json.dump(
                {fid: flight.to_dict() for fid, flight in self.flights.items()},
                f,
                indent=2
            )

    def add_flight(self, flight_number, origin, destination, total_seats):
        try:
            flight_number = flight_number.strip().upper()
            origin = origin.strip().title()
            destination = destination.strip().title()

            if not flight_number:
                raise ValueError("Flight number cannot be empty")
            if not origin or not destination:
                raise ValueError("Origin/destination cannot be empty")
            if total_seats <= 0:
                raise ValueError("Total seats must be > 0")
            if flight_number in self.flights:
                raise ValueError("Flight already exists")

            self.flights[flight_number] = Flight(
                flight_number, origin, destination, total_seats
            )
            logging.info(f"Added flight {flight_number}")
            self.save_flights()
        except ValueError as e:
            logging.error(f"Add flight failed: {str(e)}")

    def find_trip(self, origin, destination):
        try:
            origin = origin.strip().title()
            destination = destination.strip().title()

            if not origin or not destination:
                raise ValueError("Invalid search parameters")

            results = [
                flight for flight in self.flights.values()
                if flight.origin == origin
                and flight.destination == destination
            ]

            if not results:
                logging.info(f"No flights from {origin} to {destination}")
                return

            for flight in results:
                logging.info(
                    f"Flight {flight.flight_number}: "
                    f"{flight.origin} → {flight.destination} | "
                    f"Seats: {flight.available_seats}/{flight.total_seats}"
                )
        except Exception as e:
            logging.error(f"Search error: {str(e)}")

    def reserve_seat(self, flight_number, passenger_name, passenger_contact):
        try:
            flight_number = flight_number.strip().upper()
            passenger_name = passenger_name.strip().title()

            if flight_number not in self.flights:
                raise ValueError("Flight not found")

            flight = self.flights[flight_number]

            if flight.available_seats <= 0:
                raise ValueError("Flight full")

            passenger_id = f"{passenger_name.lower()}_{flight_number}"
            if passenger_id in flight.passengers:
                raise ValueError("Passenger already booked")

            flight.passengers[passenger_id] = Passenger(passenger_name, passenger_contact)
            flight.available_seats -= 1
            flight.reserved_seats += 1
            logging.info(f"Booked {passenger_name} on {flight_number}")
            self.save_flights()
        except ValueError as e:
            logging.error(f"Booking failed: {str(e)}")

    def release_seat(self, flight_number, passenger_name):
        try:
            flight_number = flight_number.strip().upper()
            passenger_name = passenger_name.strip().title()

            if flight_number not in self.flights:
                raise ValueError("Flight not found")

            flight = self.flights[flight_number]
            passenger_id = f"{passenger_name.lower()}_{flight_number}"

            if passenger_id not in flight.passengers:
                raise ValueError("No booking found")

            del flight.passengers[passenger_id]
            flight.available_seats += 1
            flight.reserved_seats -= 1
            logging.info(f"Released seat for {passenger_name}")
            self.save_flights()
        except ValueError as e:
            logging.error(f"Release failed: {str(e)}")

    def cancel_reservation(self, flight_number, passenger_name):
        self.release_seat(flight_number, passenger_name)

    def performance_report(self):
        if not self.flights:
            logging.info("No flights available")
            return

        logging.info("\n=== PERFORMANCE REPORT ===")
        for flight in self.flights.values():
            utilization = (flight.reserved_seats / flight.total_seats) * 100
            logging.info(
                f"Flight {flight.flight_number}:\n"
                f"Route: {flight.origin} → {flight.destination}\n"
                f"Capacity: {flight.total_seats} | "
                f"Booked: {flight.reserved_seats} | "
                f"Available: {flight.available_seats}\n"
                f"Utilization: {utilization:.1f}%\n"
            )

if __name__ == "__main__":
    # Clear previous data for clean testing
    open("flights.json", "w").close()

    system = FlightBookingSystem()

    print("\n=== Testing 10 Edge Cases ===")

    # 1. Add invalid flights
    print("\n1. Invalid flight additions:")
    system.add_flight("", "NY", "London", 100)     # Empty flight number
    system.add_flight("FL101", "", "Paris", 200)   # Empty origin
    system.add_flight("FL102", "Tokyo", "", 300)   # Empty destination
    system.add_flight("FL103", "Berlin", "Rome", 0)# Invalid seats

    # 2. Duplicate flight
    print("\n2. Duplicate flight:")
    system.add_flight("FL200", "Mumbai", "Delhi", 150)
    system.add_flight("fl200", "Mumbai", "Delhi", 200)  # Case duplicate

    # 3. Reserve on non-existent flight
    print("\n3. Reserve invalid flight:")
    system.reserve_seat("FL999", "Alice", "alice@test.com")

    # 4. Release on non-existent flight
    print("\n4. Release invalid flight:")
    system.release_seat("FL999", "Bob")

    # 5. Cancel on non-existent flight
    print("\n5. Cancel invalid flight:")
    system.cancel_reservation("FL999", "Charlie")

    # 6. Overbooking test
    print("\n6. Overbooking test:")
    system.add_flight("FL300", "London", "Dubai", 2)
    system.reserve_seat("FL300", "Dave", "dave@test.com")
    system.reserve_seat("FL300", "Eve", "eve@test.com")
    system.reserve_seat("FL300", "Frank", "frank@test.com")  # Should fail

    # 7. Release non-existent booking
    print("\n7. Release invalid booking:")
    system.release_seat("FL300", "Grace")

    # 8. Cancel non-existent booking
    print("\n8. Cancel invalid booking:")
    system.cancel_reservation("FL300", "Henry")

    # 9. Invalid search
    print("\n9. Invalid search:")
    system.find_trip("", "Dubai")
    system.find_trip("London", "")

    # 10. Non-existent route
    print("\n10. Non-existent route:")
    system.find_trip("Sydney", "Tokyo")

    # Valid operations demo
    print("\n=== Valid Operations Demo ===")
    system.add_flight("FL400", "Paris", "Rome", 3)
    system.reserve_seat("FL400", "Ivy", "ivy@test.com")
    system.reserve_seat("FL400", "Jack", "jack@test.com")
    system.release_seat("FL400", "Ivy")
    system.find_trip("Paris", "Rome")

    print("\n=== Final Report ===")
    system.performance_report()
```
